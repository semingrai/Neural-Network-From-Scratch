```python
inputs = [1, 2, 3, 2.5]
weights =  [[0.2, 0.8, -0.5, 1.0],
            [0.5, -0.91, 0.26, -0.5], 
            [-0.26, -0.27, 0.17, 0.87]]
biases = [2, 3, 0.5]
```

```python
# Output of current layer
layer_outputs = []

# For each neuron
for neurons_weights, neuron_bias in zip(weights, biases):
    # Zero output of given neuron
    neuron_output = 0
    # For each input and weight to the neuron
    for n_inputs, weight in zip(inputs, neurons_weights):
        #multiply by associated weight and add to the neurons output variable
        neuron_output += n_inputs*weight
    # adding bias
    neuron_output += neuron_bias

    #put neurons back into result
    layer_outputs.append(neuron_output)
```

```python
print(layer_outputs)
```

```python
#A layer of Neurons & batches of Data w/Numpy
inputs = [[1.0, 2.0, 3.0, 2.5],
          [2.0, 5.0, -1.0, 2.0],
          [-1.5, 2.7, 3.3, -0.8]]
weights = [[0.2, 0.8, -0.5, 1.0],
           [0.5, -0.91, 0.26, -0.5],
           [-0.26, -0.27, 0.17, 0.87]]
biases = [2.0, 3.0, 0.5]
outputs = np.dot(inputs, np.array(weights).T)+biases
```
