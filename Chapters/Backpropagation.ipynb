```python
x = [1.0, -2.0, 3.0] #input values
w = [-3.0, -1.0, 2.0] #weights
b = 1.0 # bias

xw0 = x[0] * w[0]
xw1 = x[1] * w[1]
xw2 = x[2] * w[2]
print(xw0, xw1, xw2)
```

```python
z = xw0 + xw1 + xw2 + b 
z
```

```python
y = max(z,0) #ReLU activation function
y
```

```python
dvalue = 1.0 # since the output here is 6 it's derivative is 1 for this example
drelu_dz = dvalue * (1. if z>0 else 0.)
drelu_dz
```

```python
dsum_dxw0 = 1 # partial derivative of x respect to the x input weight for 0th pair of inputs and weight
drelu_dxw0 = drelu_dz * dsum_dxw0
drelu_dxw0
```

```python
dsum_dxw1 = 1 # partial derivative of x respect to the x input weight for 0th pair of inputs and weight
drelu_dxw1 = drelu_dz * dsum_dxw1
drelu_dxw1
```

```python
dsum_dxw2 = 1 # partial derivative of x respect to the x input weight for 0th pair of inputs and weight
drelu_dxw2 = drelu_dz * dsum_dxw2
drelu_dxw2
```

```python

dsum_db = 1 # biases 
drelu_db = drelu_dz * dsum_db
drelu_db
```

```python
# Back propagation/derivative of ReLU(sum(wx+b))
# where ReLU 0 or 1
# (drelu/dsum) * (dsum/dxw and b also) * (dxw/dx and dxw/dw(partial derivative))
```

```python
dmul_dx0 = w[0]
dmul_dw0 = x[0]
drelu_dx0 = drelu_dxw0 * dmul_dx0
drelu_dw0 = drelu_dxw0 * dmul_dw0
drelu_dx0
```

```python
dmul_dx1 = w[1]
dmul_dw1 = x[1]
drelu_dx1 = drelu_dxw1 * dmul_dx1
drelu_dw1 = drelu_dxw1 * dmul_dw1
drelu_dx1
```

```python
dmul_dx2 = w[2]
dmul_dw2 = x[2]
drelu_dx2 = drelu_dxw2 * dmul_dx2
drelu_dw2 = drelu_dxw2 * dmul_dw2
drelu_dx2
```

```python
print(drelu_dx0, drelu_dw0, drelu_dx1, drelu_dw1, drelu_dx2, drelu_dw2)
```

```python
drelu_dx0 = dvalue * (1. if z > 0 else 0.) * w[0]
drelu_dx1 = dvalue * (1. if z > 0 else 0.) * w[1]
drelu_dx2 = dvalue * (1. if z > 0 else 0.) * w[2]
print(drelu_dx0, drelu_dx1, drelu_dx2)
```

```python
dx = [drelu_dx0, drelu_dx1, drelu_dx2] # gradients on inputs
dw = [drelu_dw0, drelu_dw1, drelu_dw2] # gradients on weights
db = drelu_db # gradient on bias... just 1 bias here
```

```python
print(w, b)
```

```python
w[0] += -0.001 * dw[0]
w[1] += -0.001 * dw[1]
w[2] += -0.001 * dw[2]
b += -0.001 * db
```

```python
print(w,b)
```

```python
xw0 = x[0] * w[0]
xw1 = x[1] * w[1]
xw2 = x[2] * w[2]
z = xw0 + xw1 +xw2 +b
y = max(z,0)
print(y)
```
