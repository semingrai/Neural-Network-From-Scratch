```python
import numpy as np
```

```python
layer_outputs = [4.8, 1.21, 2.385]
#First step is to "exponentiate" the outputs-Euler's number e, E to match common coding
E = 2.71828182846

# for each value in a vector, calculate the exponential value

exp_values = []
for output in layer_outputs:
    exp_values.append(E**output)
exp_values
```

```python
# sum of exponentiated values
sUm = sum(exp_values)
# Sij or normalized values
Sij = []
for i in exp_values:
    Sij.append(i/sUm)
```

```python
norm_values = Sij
print(f"Normalised values: {norm_values}")
print(f"Sum of Normalized values: {sum(norm_values)}")
```

```python
# Numpy way
layer_outputs = [4.8, 1.21, 2.385]
exp_values = np.exp(layer_outputs)
print(f"Exponentiated values: {exp_values}")

#normalized values
norm_values = exp_values / np.sum(exp_values)
print(f"Normalised values: {norm_values}")
print(f"Sum of Normalized values: {np.sum(norm_values)}")
```

```python
# Get unnormalized probabilities
exp_values = np.exp(norm_values)
probablities = exp_values / np.sum(layer_outputs, axis=None, keepdims=True)
```

```python
#Softmax Activation
class Activation_Softmax:
    # Forward pass
    def forward(self, inputs):
        # Get unnormalized probabilities
        exp_values = np.exp(inputs - np.max(inputs, axis=1, keepdims=True))

        # Normalize them for each sample
        probablities = exp_values / np.sum(exp_values, axis=1, keepdims=True)
        self.output = probablities
```

```python
softmax = Activation_Softmax()
softmax.forward([[1,2,3]])
print(softmax.output)
```

```python
import matplotlib.pyplot as plt

iterations = list(range(1, 11))  # 10 iterations for example
# Loss values for 3 learning rates (simulated)
loss_lr_small = [10, 9, 8, 7.5, 7, 6.5, 6.2, 6, 5.8, 5.7]    # slow
loss_lr_medium = [10, 7, 4, 2.5, 1.5, 0.8, 0.4, 0.2, 0.1, 0.05] # fast
loss_lr_large = [10, 12, 15, 14, 16, 18, 20, 25, 30, 35]        # diverging

plt.plot(iterations, loss_lr_small, label='lr=0.001 (slow)')
plt.plot(iterations, loss_lr_medium, label='lr=0.01 (fast)')
plt.plot(iterations, loss_lr_large, label='lr=0.1 (diverges)')

plt.xlabel('Iterations')
plt.ylabel('Loss')
plt.title('Gradient Descent Convergence for Different Learning Rates')
plt.legend()
plt.show()
```
